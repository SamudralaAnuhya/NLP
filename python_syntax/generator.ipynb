```python
#merging and joing df 
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2',  'A3'],
                    'B': ['B0', 'B1', 'B2','B3'],
                    'C': ['C0', 'C1', 'C2', 'C3']})

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6',  'A7'],
                    'B': ['B4', 'B5', 'B6','B7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})

pd.merge(df1, df2, on = 'B', how = 'outer')  
#outer join , left join, right join, inner join,lefe outer join, right outer join
```
