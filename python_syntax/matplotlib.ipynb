```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('-axis')
plt.ylabel('y-axis')
plt.title('My first plot')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]


plt.plot(x, y, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('-axis')
plt.ylabel('y-axis')
plt.title('My first plot')
plt.grid()
plt.show()
```

```python
#multiple plots
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(4,2))   
plt.subplot(1, 2, 1)  #we are constructing a 1x6 grid means 6 graphs in line 1  and 3rd cordinate is at which position we are drawing
plt.plot(x, y ,'g')

plt.subplot(1, 2, 2)  #we are constructing a 1x6 grid of plots and we are selecting the second plot
plt.plot(x, y2, 'r')
plt.show()


```

```python
#barchart 
categories = ['A', 'B', 'C', 'D', 'E']
y = [1, 4, 9, 16, 25]

plt.bar(categories, y ,color = 'purple')
```

```python
#pie chart 
sizes = [25, 20, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightblue']
explode = (0.2, 0, 0, 0)  # explode 1st slice
plt.pie(sizes, explode= explode , labels=labels, colors=colors, autopct='%1.1f%%' , shadow=True)
```
